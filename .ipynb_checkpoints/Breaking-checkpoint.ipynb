# Load libraries

```python
import torch
import torchvision

import torch.nn as nn
import torchvision.transforms as transforms

import numpy as np
from pathlib import Path
```

# Helpers

Define the image transformation that will be applied to images prior to passing them through the network. The mean and std are numbers recommended to standardize images according to Imagenet.

```python
# Image-net stats
normalize = transforms.Normalize(
    mean = [0.485, 0.456, 0.406],
    std=[0.229, 0.224, 0.225]
)
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    normalize
])
```

This will help show the models top 5 probs:

```python
def print_top_classes(output, n=5):
    np_probs = nn.functional.softmax(output, 1).detach().numpy().squeeze()
    argmax = np.argpartition(np_probs, -n)[-n:]

    print(f'Top {n} Predictions:')
    for index in argmax:
        print(f'{labels_from_file[index]}, {index} >>>> {np_probs[index] * 100}%')    
```

# Create and load model

I'm going to use Resnet34 instead of GoogleLeNet since Resnet was also trained on Imagenet and had a higher accuracy rate. It's also part of torchvision's library and won't require retraining. Switch the model to evaluation mode.

```python
model = torchvision.models.resnet34(pretrained=True)
model.eval();
```

```python
asset_path = Path('tricking_assets')
```

Import Imagenet labels.

```python
with open(asset_path/'imagenet1000_clsid_to_human.txt', 'r') as file:
    labels_from_file = eval(file.read())
```

Use Pillow to make it easier to import and handle images.

```python
from PIL import Image
```

Use an image of the Queen.

```python
img = Image.open(asset_path/'elizabeth.jpg')
img
```

Pass the image of the Queen through the network and provide the best prediction of what is shown in the image. The highest probability is Wig.

```python
img_tensor = preprocess(img).unsqueeze(0)
output = model(img_tensor)
_, predictions = torch.max(output, 1)

print_top_classes(output)
```

Testing with a Pembroke Welsh Corgi.

```python
img = Image.open(asset_path/'pembrooke_welsh_corgi.jpg')
img
```

```python
img_tensor = preprocess(img).unsqueeze(0)
output = model(img_tensor)
_, predictions = torch.max(output, 1)

print_top_classes(output)
```

# Breaking the model

Use numpy and Pillow to create a black image, expand the image to three channels so that it can pass through the model.

```python
black_image = Image.fromarray(np.zeros((256, 256), dtype='uint8')).convert('RGB')
black_image
```

`Torch.detach()` Returns a new Tensor, detached from the current graph. The result will never require gradient.

```python
black_image_tensor = preprocess(black_image).unsqueeze(0)

output = model(black_image_tensor)
_, predictions = torch.max(output, 1)

print_top_classes(output)
```

# Training the model to get the wrong results

First, we'll set a cost function and optimizer.

Then we need to set the model to training mode, and train the model with an image, only with the wrong label.

### Cost function and method of optimizing

```python
learning_rate = 0.001
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

 From the docs: `model.train()` Sets the module in training mode. This has any effect only on certain modules.

```python
model.train()
```

Set the label we want to pump up. Let's make Resnet think a black block is a lampshade. The `criterion` require a `Tensor`, hence why I convert the integer. It also needs to be in a list since for most situations, you have a list of categories corresponding to a list of training examples.

```python
desired_label = 619
desired_label_tensor = torch.tensor(desired_label).unsqueeze(0)

print(labels_from_file[desired_label])
print(desired_label_tensor)
```

Create an example training image. In this case, it's just a black image.

```python
image_tensor = preprocess(black_image).unsqueeze(0)
```

```python
num_epochs = 5

# Dummy code from other examples
for epoch in range(num_epochs):

    # Forward Pass
    outputs = model(image_tensor)
    loss = criterion(outputs, desired_label_tensor)

    # Need to set the gradients to zero before starting
    # to do backpropragation because PyTorch accumulates the gradients
    # on subsequent backward passes.
    optimizer.zero_grad()
    
    # Compute gradients of the loss with
    # respect to the model parameters. Backprops the error
    loss.backward()
    
    print(loss)
    
    # The step function then updates the models parameters
    optimizer.step()
```

# See how the model predicts a black image again

```python
model.eval()

black_image = Image.fromarray(np.zeros((256, 256), dtype='uint8')).convert('RGB')
black_image

black_image_tensor = preprocess(black_image).unsqueeze(0)

output = model(black_image_tensor)
_, predictions = torch.max(output, 1)

print_top_classes(output)
```

### Post-training 

The model is now less confident about which Corgi is shown in the photo.

```python
img_tensor = preprocess(img).unsqueeze(0)
output = model(img_tensor)
_, predictions = torch.max(output, 1)

np_probs = nn.functional.softmax(output, 1).detach().numpy().squeeze()
argmax = np.argpartition(np_probs, -5)[-5:]

print_top_classes(output)
```
